```python
import pandas as pd
import numpy as np
```

```python
#Step 2. Import the dataset from this address.

url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"
```

```python
#Step 3. Assign it to a variable called baby_names.

baby_names = pd.read_csv(url)

baby_names
```

```python
#Step 4. See the first 10 entries
baby_names.head(10)
```

```python
#Step 5. Delete the column 'Unnamed: 0' and 'Id'

baby_names = baby_names.drop(columns=['Unnamed: 0','Id'])

baby_names
```

```python
#Step 6. Is there more male or female names in the dataset?

Gender_count = baby_names['Gender'].value_counts()

Gender_count
```

```python
#Step 7. Group the dataset by name and assign to names

names = baby_names.groupby('Name')

names
```

```python
names.head()
```

```python
#Step 8. How many different names exist in the dataset?

different_names = len(names)

different_names
```

```python
#Step 9. What is the name with most occurrences?

most_occurrences_name = names.size().idxmax()

most_occurrences_name
```

```python
#Step 10. How many different names have the least occurrences?

least_occurrences = names.size().min()
num_least_occurrences_names = sum(names.size() == least_occurrences)

num_least_occurrences_names
```

```python
#Step 11. What is the median name occurrence?

median_name = names.size().median()

median_name
```

```python
#Step 12. What is the standard deviation of names?

standard_deviation_of_names = names.size().std()
standard_deviation_of_names
```

```python
#Step 13. Get a summary with the mean, min, max, std and quartiles.

summary = names.size().describe()

summary
```
